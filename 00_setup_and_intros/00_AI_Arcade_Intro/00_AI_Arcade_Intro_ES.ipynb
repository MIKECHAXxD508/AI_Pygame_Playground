### Aprendizaje por Refuerzo

¿Qué es el Aprendizaje por Refuerzo? Lo veremos más adelante, pero en resumen, es un tipo de aprendizaje que no requiere que des ninguna instrucción al algoritmo. En su lugar, el algoritmo (o Agente) que crearás aprenderá mediante prueba y error (y recompensas) a navegar por el entorno en el que se encuentra. Cuando el Agente toma una decisión correcta (es decir, alcanza la meta o un subobjetivo) recibirá una recompensa (digamos un +1). A lo largo de sucesivas iteraciones, el Agente aprenderá lo suficiente sobre su entorno como para navegar con éxito incluso cuando se añada aleatoriedad.

### ¿Por qué deberías aprender sobre Aprendizaje por Refuerzo?

Deberías aprender sobre aprendizaje por refuerzo (RL) porque te capacita para diseñar sistemas que pueden mejorar automáticamente con el tiempo, adaptarse a nuevas situaciones y resolver problemas complejos de toma de decisiones. Dominar RL puede llevar a avances en diversos campos, permitiendo innovaciones que los métodos tradicionales no pueden lograr.

- Robótica: Enseña a los robots a realizar tareas de forma autónoma.
- IA para juegos: Crea agentes inteligentes para dominar juegos complejos.
- Vehículos autónomos: Desarrolla la toma de decisiones en tiempo real para coches autónomos.
- Trading financiero: Optimiza estrategias de trading para obtener mayores rendimientos.
- Atención médica: Personaliza tratamientos y optimiza resultados médicos.
- Investigación operativa: Mejora la gestión de la cadena de suministro y recursos.

# Estructura del Curso

### Enfoque en lo Práctico

Este curso está diseñado para darte una idea de cómo es ser un ingeniero de IA en la práctica. En lugar de profundizar en términos complejos o ecuaciones, nos centraremos en lo práctico.

### Comenzando con lo Básico

Este curso asume que todos son principiantes en el campo de la programación en general (así como en el Aprendizaje Automático). Por lo tanto, nuestro primer módulo cubrirá estrictamente los conceptos básicos de Python (el lenguaje de programación número uno para el Aprendizaje Automático) y cómo usar Git (un marco para almacenar tu código y colaborar con otros).

Después de cubrir estos temas fundamentales, también necesitaremos ver el marco que usaremos para interactuar con nuestros mundos de videojuegos. Este marco se llama Gymnasium, y esencialmente se utiliza para empaquetar entornos de videojuegos con un conjunto de reglas o funciones estandarizadas que facilitan el entrenamiento de algoritmos de aprendizaje automático.

### Aprendizaje Q (Quality Learning) y Aprendizaje Q con Redes Neuronales

El primer algoritmo de aprendizaje automático que aprenderemos será el Aprendizaje Q o Quality Learning. Este es un algoritmo muy simple pero altamente eficiente que funciona bien cuando se aplica a juegos de baja complejidad.

En segundo lugar, aprenderemos cómo agregar Redes Neuronales Artificiales (el cerebro de nuestro Agente) al Aprendizaje Q, lo que nos dará la capacidad de aplicar nuestro algoritmo a juegos más complejos.

### Proyecto Final

Para nuestro proyecto final nos dividiremos en grupos. Un grupo trabajará en la construcción de un entorno de juego único, el segundo trabajará en la construcción de un algoritmo de aprendizaje automático para jugar el juego, y el último diseñará los activos pixelizados (personajes, escenarios, etc.) para el entorno del juego.
