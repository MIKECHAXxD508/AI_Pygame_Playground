### Renforcement Learning

What is Renforcement Learning? We'll get to that, but in short it's a type of learning that deosn't require you to give any instructions to the algorythm. Instead the algorythim (or Agent) that you will create will learn through trial and error (and reward) how to navigate the environment it's placed in. When the Agent makes a correct decision (i.e. reaches the goal or sub goal) it will recieve a reward (let's say a +1). Over sucessive iterations the Agent learns enough about it's environment that I can navigate sucessfully even when randomness is added.

### Why should you learn about Renforcement Learning?

Reinforcement learning (RL) should be learned because it equips you with the ability to design systems that can automatically improve over time, adapt to new situations, and solve complex decision-making problems. Mastering RL can lead to breakthroughs in various fields, enabling innovations that traditional methods cannot achieve.

- Robotics: Teach robots to perform tasks autonomously.
- Game AI: Create intelligent agents for mastering complex games.
- Autonomous Vehicles: Develop real-time decision-making for self-driving cars.
- Financial Trading: Optimize trading strategies for higher returns.
- Healthcare: Personalize treatments and optimize medical outcomes.
- Operations Research: Enhance supply chain and resource management.

# Course Structure

### Focus on the Practical

This course is meant to give you an idea of what it's like to be an AI engineer in practice. rather than go into depth on complex terms or equations we will focus on the practical.

### Starting off with the Basics

This course assumes that everyone is a beginner to the field of programming in general (as well as Machine Learning). Therefore, our first module will strictly cover the basics of Python (the number one programming langauge for Machine Learning) and how to use git (a framework for storing your code, and collaborating with others).

After these foundational topics are covered we'll also need to look at the framework we'll be using to interact with our video game worlds. This framework is called Gymnasium, and it is essentially used to package up video game environments with a set of standardized rules or functions which makes it easy to train machine learning algorythms on.

### Q learning (Quality Learning) and Q Learning with Neural Networks

The first machine learning algorythm we will learn will be Q learning or Quality Learning. This is a very simple yet highly proformant algorythm that works well when applied to low complexity games.

Secondly we will learn how to add Artifical Neural Networks (the brian of our Agent) to Q learning giving us the ability to apply our algorythm to more complex games.

### Final Project

for our final project we will break up into groups. one group will work building a unique game environment, the second will work on building a maching learning algorythm to play the game, and the last will design the pixalized assets (characters, scenery, etc.) for the game environment.
